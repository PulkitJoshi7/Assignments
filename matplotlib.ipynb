Matplotlib is an amazing visualization library in Python for 2D plots of arrays. 
Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.
five plots that can be plotted are as follows:
Bar
Line
Scatter
stem
spectogram

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])




# fig = plt.figure()
# fig11=fig.add_subplot()
# fig12=fig.add_subplot()
# fig13=fig.add_subplot()
# fig14=fig.add_subplot()
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)

plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(2,2,1)
plt.bar(company,profit)
plt.subplot(2,2,3)
plt.barh(company, profit)
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plotdata=[box1,box2]

plt.boxplot(plotdata,notch='True',patch_artist=True,labels=['Box1','Box2'])
```
